```python
def knapSack01(W, wt, val, n):
  
    if n == 0 or W == 0 :
        return 0
  
    if (wt[n-1] > W):
        return knapSack01(W, wt, val, n-1)
  
    else:
        return max(val[n-1] + knapSack01(W-wt[n-1], wt, val, n-1),
                   knapSack01(W, wt, val, n-1))
 
```

```python
def BubbleSort(lst,lst2,lst3):
    
    for i in range(len(lst)):
        for j in range(len(lst)-1):
            if lst[j] > lst[j+1]:
                lst[j] , lst[j+1] = lst[j+1] , lst[j]
                lst2[j] , lst2[j+1] = lst2[j+1] , lst2[j]
                lst3[j] , lst3[j+1] = lst3[j+1] , lst3[j]
                
    return lst,lst2,lst3
```

```python
def OrderChoise(val,wt,order):
    lst3=[0]*len(val)
    if order == "min weight":
        lst=wt
        lst2=val
        lst,lst2,lst3=BubbleSort(lst,lst2,lst3)
        
        return lst2,lst
        
    elif order == "max profit":
        lst=val
        lst2=wt
        lst,lst2,lst3=BubbleSort(lst,lst2,lst3)
        lst2.reverse()
        lst.reverse()
       
        return lst,lst2
        
    elif order == "max profit/weight":
        lst=[i/j for i,j in list(zip(val,wt))]
        lst2=val
        lst3=wt
        lst,lst2,lst3=BubbleSort(lst,lst2,lst3)
        lst3.reverse()
        lst2.reverse()
       
        return lst2,lst3
```

```python
def knapSackFrac(Bag_Weight, weight, value):
    val_wt = list(zip(value,weight))
    print(val_wt)
    profit = 0
    remainder = Bag_Weight
    for p,w in val_wt:
         if remainder < w:
            p = p*remainder/w
            profit += p
            return profit
         else:
            remainder = remainder - w
            profit += p
```

```python
def AllKnabsackFrac(W,val, wt):
     val1,wt1 = OrderChoise(val,wt,"max profit")
   
     profit=knapSackFrac(W, wt1,val1)  
     print("**************profit at max values = " +str(profit)+"**************")
     val2,wt2 = OrderChoise(val,wt,"min weight")
     profit2=knapSackFrac(W, wt2,val2)       
     print("**************profit at min weights = " +str(profit2)+"**************")
     val3,wt3 = OrderChoise(val,wt,"max profit/weight")
     profit3=knapSackFrac(W, wt3,val3)
              
     print("**************profit at max value/weight= " +str(profit3)+"**************")  
```
